## Matplotlib-Subplots

### Functional Method

```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
x = np.linspace(0,10,10)
print(x)
```

```python
x = np.linspace(0,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y1,'ro') # r: red, o: circle
plt.title('$y_1=x$')
# "_" will push the index lower right of the variable.
# If we put $ to both sides we will push the power to upper right.

plt.subplot(2,2,2)
plt.plot(x,y2,'g--') # g: green, --: dash
plt.title('$y_2=x^2$') 

plt.subplot(2,2,3)
plt.plot(x,y3,'b^') # b: blue, ^: triangle
plt.title('$y_3=x^3$')

plt.subplot(2,2,4)
plt.plot(x,y4,'ks') # k: black, s: square
plt.title('$y_4=\sqrt{x}$')
```

#### Another way of Rendering

```python
x = np.linspace(0,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()

plt.subplot(4,1,1)
plt.plot(x,y1,'ro') # r: red, o: circle
plt.title('$y_1=x$')

plt.subplot(4,1,2)
plt.plot(x,y2,'g--') # g: green, --: dash
plt.title('$y_2=x^2$') 

plt.subplot(4,1,3)
plt.plot(x,y3,'b^') # b: blue, ^: triangle
plt.title('$y_3=x^3$')

plt.subplot(4,1,4)
plt.plot(x,y4,'ks') # k: black, s: square
plt.title('$y_4=\sqrt{x}$')
```

### Histogram
Used with interva/ratio variables.\
Represent the frequency of each attribute for a variable.\
Good overview of the distribution of our data.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins = 3, rwidth = 0.95) 
# "bins" are nothing but ranges in histogram.
# rwidth: relative witdth of a bar compared to a bin size. 
```

### Box Plot
A box plot is a method for graphically depocting groups of numerical data through their quertiles.

#### A box plot is a way of statistically representing the distribution of the data through five main dimensions:
 - Minimum: The smallest number in the dataset excluding the outliers.
 - First quartile: Middle number between the minimum and the median.
 - Second quartile (Median): Middle number of the (sorted) dataset.
 - Third quartile: Middle number between median and maximum.
 - Maximum: The largest number in the dataset excluding the outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

x_bp = np.random.normal(170,10,250)

plt.boxplot(x_bp)
plt.show()
```

### Pie Chart
 - Often used with nominal and ordinal variables.
 - Circle cut into "pie slices" that add up to 100%.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1900,605,385,200,140,160,1000]
exp_labels = ['Rent','Honda','VW','PG&E','Water','Tel/Int','food']
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, explode=[0.1,.1,0,0,0,0,0], radius=2,
        autopct="%.1f") # autopct="%.1f" will work too.
plt.show()
```

```python
my_explode = [0.1,.1,0,0,0,0,0]
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, explode=my_explode, radius=2,
        autopct="%.1f%%", shadow=True, startangle=180) # autopct="%.1f" will work too.
plt.show()
```
