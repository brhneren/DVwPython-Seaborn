# Seaborn: scatterplot
Scatterplot is considered a relational plot.

```python
seaborn.scatterplot(x=None, y=None, hue=None, data=None,  ci=95, legend='auto')
```

````python
import seaborn as sns
from matplotlib import pyplot as plt
````

````python
diamonds = sns.load_dataset('diamonds') # This data is in the seaborn liabrary itself.

diamonds.shape
````

````python
diamonds.head()
````

````python
diamonds = diamonds[
    diamonds.cut.isin(['Premium','Good']) & 
    diamonds.color.isin(['D','F','J'])
].sample(n=100, random_state=22)
````

````python
diamonds.shape
````

## Intro Visuals

## Basics

````python
sns.set_style('dark')
````

````python
sns.scatterplot(diamonds.carat, diamonds.price);
````

````python
sns.scatterplot(x='carat', y='price', data=diamonds);
# Another way to do the previous plot.  # this will bring the entire dataframe
````

````python
sns.scatterplot(x='x', y='y', data=diamonds);
# Shows a strong relationship between x and y values.
````

## Semantic Variables

````python
sns.set_style('darkgrid')
````

````python
diamonds = diamonds[
    diamonds.cut.isin(['Premium','Good']) & 
    diamonds.color.isin(['D','F','J'])
].sample(n=100, random_state=22)
````

### hue

````python
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds);
````

````python
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds,
                palette=['purple', '#55CCCC', 'blue', 'red', 'green']);

# #55CCCC = Acqua
````

### Continues Variable

````python
sns.scatterplot(x='carat', y='price', hue='depth', data=diamonds);
````

### Style

````python
sns.scatterplot(x='carat', y='price', hue='cut', style='cut', data=diamonds,
                s=100); # s: controls the size of the dots.
````

### Size

````python
sns.scatterplot(x='carat', y='price', hue='cut', size='cut', data=diamonds);
````

````python
sns.scatterplot(x='carat', y='price', hue='cut', size='cut', data=diamonds,
               sizes=[100,150,200,125,50],
                palette=['purple', '#55CCCC', 'blue', 'red', 'green']);
````

````python
sns.scatterplot(x='carat', y='price', size='depth', data=diamonds);
````

### hue + style + size

````python
sns.scatterplot(x='carat', y='price',
                hue='cut',
                size='cut', data=diamonds,
                style='cut',
                palette=['purple', '#55CCCC', 'blue', 'red', 'green'],
                s=100);
````

````python
sns.scatterplot(x='carat', y='price',
                hue='cut',
                size='cut', data=diamonds,
                style='color',
                palette=['purple', '#55CCCC', 'blue', 'red', 'green'],
                s=150);
````

## Additional Styling

### Alpha
Is for transparency

````python
sns.scatterplot(x='carat', y='price',
                hue='cut',
                size='cut', data=diamonds,
                style='color',
                palette=['purple', '#55CCCC', 'blue', 'red', 'green'],
                s=150,
                alpha=0.6);
````

### Marker Styling

Check the matplotlib scatter and the marker style docs for more styling of the markers themselves

````python
sns.scatterplot(x='carat', y='price', 
                data=diamonds, 
                s=200,
                marker='*');
````

````python
sns.scatterplot(x='carat', y='price', 
                data=diamonds, 
                s=200,
                marker='*',
                edgecolor='black'
               );
````
