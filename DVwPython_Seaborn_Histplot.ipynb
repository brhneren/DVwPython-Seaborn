````python
import seaborn as sns
from matplotlib import pyplot as plt
````

````python
penguins=sns.load_dataset('penguins')
````

````python
penguins.head()
````

````python
penguins.shape
````

````python
penguins.dropna(inplace=True)
````

````python
penguins.shape
````

## Intro Visuals

## Basics

````python
sns.set_style('darkgrid')
````

````python
sns.histplot(penguins.bill_length_mm);
````

````python
sns.histplot(x='bill_length_mm', data=penguins);
````

````python
sns.histplot(y='bill_length_mm', data=penguins);
````

### kde

````python
sns.histplot(x='bill_length_mm', data=penguins,
            kde = True);
````

### bins

````python
sns.histplot(x='bill_length_mm', data=penguins,
            bins=20);
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            bins=[30,40,45,50,55,60]);
````

### binwidth and binrange

````python
sns.histplot(x='bill_length_mm', data=penguins,
            binwidth=10, binrange=[30,60]);
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            binwidth=5, binrange=[30,60]);
````

## Statistics

### stat

````python
sns.histplot(x='bill_length_mm', data=penguins,
            stat='count');
````

### density
AREA of histogram sums to one

````python
sns.histplot(x='bill_length_mm', data=penguins,
            stat='density');
````

### probability
HEIGHT of histogram bars to one

````python
sns.histplot(x='bill_length_mm', data=penguins,
            stat='probability');
````

### cumulative

````python
sns.histplot(x='bill_length_mm', data=penguins,
            stat='probability',
             fill=False,
             element='step',
             cumulative=False,);
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            stat='probability',
             fill=False,
             element='step',
             cumulative=True,);
````

## Categorical Variables

### hue

````python
penguins.head()
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species');
````

### element

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            element='step');
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            element='poly'); # produces ploygons
````

### multiple

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            multiple='layer'); # layer is the default value 
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            multiple='stack');
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            multiple='fill');
````

### BONUS Example: ```multiple="dodge"```

````python
sns.histplot(x='bill_length_mm', data=penguins,
            hue='species',
            multiple='dodge');
````

## Bivariate Histplot

````python
sns.set_style('white')
````

````python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins);
# We need to pass both x and y variable into the code.
# This create a heat map
````

### cbar

````python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
            cbar=True); # cbar is for column bar
````

### hue

````python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins,
            hue='species');
````

### Discrete color map

````python
sns.histplot(x='bill_length_mm', y='species', data=penguins,
            hue='species');
````

## Styling

### ```shrink```

````python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins);
````

````python
sns.histplot(x='species',  hue='sex', multiple='dodge', data=penguins,
             shrink=0.8);
# shrink is good with 'dodge' style.
````

### ```color```, ```palette```

````python
sns.histplot(x='bill_length_mm', data=penguins,
             color='indigo');
````

````python
sns.histplot(x='species', hue='sex', multiple='dodge', data=penguins,
             shrink=0.8, palette='bone');
````

### ```fill```

````python
sns.histplot(x='bill_length_mm', data=penguins,
             fill=True);
````

````python
sns.histplot(x='bill_length_mm', data=penguins,
             fill=False);
````

## Other ```kwargs```

Other keyword arguments can also be used, but allowed keywords depend on the type of histplot you are making:

- Univariate, element="bars": [matplotlib.axes.Axes.bar()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html)
- Univariate, other element, fill=True: [matplotlib.axes.Axes.fill_between()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.fill_between.html)
- Univariate, other element, fill=False: [matplotlib.axes.Axes.plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html)
- Bivariate: [matplotlib.axes.Axes.pcolormesh()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pcolormesh.html)
